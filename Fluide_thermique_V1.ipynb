### Théorie
 
$$ \Delta H = 0  $$

#### Quantité de chaleur cédée par le corps chaud

$$ Q_c = m_1 . c_{métal} (T_{finale} - T_{initiale})  $$

On obtient une équation avec $c_{métal}$ inconnue ($Q_c < 0 $)


#### Quantitée de chaleur absorbée par le corps froid

$$ Q_f = m_2 . c_{eau} \Delta T_{eau}   $$

données :

$$ c_{eau}\,=\,4185\,J.kg.K^{-1} $$

$m_2$ la masse d'eau dans le calorimètre.

#### Ennoncé de l'équilibre thermique 

$$ Q_f + Q_c = 0 $$

##  Simulation de la distribution de température en fonction du temps

```python
%matplotlib inline
```

```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np


import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator

# Importer cette librairie pour des versions
# de matplotlib anterieures
from mpl_toolkits.mplot3d import axes3d
```

```python
fig = plt.figure()
fig.set_dpi(100)

x = np.linspace(-5,5,30)
y = np.linspace(-5,5,30)

X,Y = np.meshgrid(x,y)


#Temperature initiale au centre à (0,0) à t=0
T = 100

# Ecart type
# Permet de paramètrer la conduction 
s = 5
```

```python
#Repartition de Temperature (données)
def u(x,y,t):
    return (T/sqrt(1+4*t/s))*exp(-(x**2+y**2)/(s+4*t))
```

```python
# Préparer les données
X = np.arange(-5, 5, 0.5)
Y = np.arange(-5, 5, 0.5)
X, Y = np.meshgrid(X, Y)

def diffusion_temp(t):


    R = (T/np.sqrt(1+4*t/s))*np.exp(-(X**2+Y**2)/(s+4*t))

    # Affichage d'une surface
    fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

    surf = ax.plot_surface(X, Y, R, cmap=cm.coolwarm,
                           linewidth=0, antialiased=False)

    # Customize the z axis.
    ax.set_zlim( 0, np.max(R) )
    ax.zaxis.set_major_locator(LinearLocator(10))

    # Une legende permet de savoir quelle est la température.
    fig.colorbar(surf, shrink=0.5, aspect=5)

    plt.show()

    
diffusion_temp(100)
```

```python
interactive(diffusion_temp, t=(0,90,1))
```
